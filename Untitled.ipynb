# Stock prediction case study


Here I used realtime stock data to predict just the closing price 

I downloaded the 'airtel stock price NSE' data from Yahoo finance

```python
import warnings
warnings.filterwarnings("ignore")

from csv import reader
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
import ta as ta
```

#### Airtel's NSE data  (2002-2019)

```python
# using airtel stock from 1-7-2002 to 24-9-2019
stock = pd.read_csv('BHARTIARTL.NS.csv')
print(stock.columns)
print('total data shape before preprocessing',stock.shape)
```

# Data PreProcessing and feature engineering

```python
# dropping all null values
stock_null = stock.dropna(axis = 0, how ='any') 
print('shape of data after removing nan',stock_null.shape)
stock_null.head()
```

```python
#Randomly plotting FIRST 300 days data 

plt.figure(figsize=(10,8))
plt.plot(np.arange(300),stock_null.Close[0:300] ) 
plt.xlabel('days') 
plt.ylabel('Adj Close') 
plt.title('Stock performance') 
plt.show() 
```

```python
# we want to predict stock prices of tommorrow using todays data so shifting close prices by one
y_close=stock_null['Adj Close'].shift(-1)
y_close.shape
```

### Splitting data according to time

```python
# splitting stock based on time
from sklearn.model_selection import train_test_split

x_train, x_test= np.split(stock_null, [int(.67 *len(stock_null))])
y_train, y_test= np.split(y_close, [int(.67 *len(y_close))])
print('shape of train:',x_train.shape)
print('shape of test:',x_test.shape)
```

```python
# dropping date from data 
x_train=x_train.drop(['Date'],axis=1)
x_test=x_test.drop(['Date'],axis=1)
x_train.head()
```

### Adding technical indicators 

The reasons for adding these indications is that these are trend indicators where most technical analystics relay on


1.Moving average of different windows

2.Exponential Moving average of different windows

3.Moving average convergence/divergence

```python
# Moving average on close price with different window sizes
x_train['MA3'] =x_train['Adj Close'].rolling(window=3).mean()
x_test['MA3'] =x_test['Adj Close'].rolling(window=3).mean()
x_train['MA2'] =x_train['Adj Close'].rolling(window=2).mean()
x_test['MA2'] =x_test['Adj Close'].rolling(window=2).mean()
 
# Moving average on open price with different window sizes
x_train['MAo3'] =x_train['Open'].rolling(window=3).mean()
x_test['MAo3'] =x_test['Open'].rolling(window=3).mean()
x_train['MAo2'] =x_train['Open'].rolling(window=2).mean()
x_test['MAo2'] =x_test['Open'].rolling(window=2).mean()

# Exponetial moving average on close price with different window sizes

x_train['ema10'] =x_train['Adj Close'].ewm(span=10, adjust=False).mean()
x_test['ema10'] =x_test['Adj Close'].ewm(span=10, adjust=False).mean()
x_train['ema5'] =x_train['Adj Close'].ewm(span=5, adjust=False).mean()
x_test['ema5'] =x_test['Adj Close'].ewm(span=5, adjust=False).mean()
x_train['ema1'] =x_train['Adj Close'].ewm(span=1, adjust=False).mean()
x_test['ema1'] =x_test['Adj Close'].ewm(span=1, adjust=False).mean()

# Exponetial moving average on open price with different window sizes
x_train['emao10'] =x_train['Open'].ewm(span=10, adjust=False).mean()
x_test['emao10'] =x_test['Open'].ewm(span=10, adjust=False).mean()
x_train['emao5'] =x_train['Open'].ewm(span=5, adjust=False).mean()
x_test['emao5'] =x_test['Open'].ewm(span=5, adjust=False).mean()
x_train['emao1'] =x_train['Open'].ewm(span=1, adjust=False).mean()
x_test['emao1'] =x_test['Open'].ewm(span=1, adjust=False).mean()

```

```python

plt.figure(figsize=(20,8))
plt.plot(np.arange(50),y_train[0:50],label='pred')
plt.plot(np.arange(50),x_train['MA3'][0:50],label='Moving average of window3' )
plt.plot(np.arange(50),x_train['ema5'][0:50],label='exponential moving average of window 5' )

# naming the x axis 
plt.xlabel('days') 
# naming the y axis 
plt.ylabel('Adj Close') 
plt.gca().legend()
  
# giving a title to my graph 
plt.title('plot for 50 days') 
  
# function to show the plot 
plt.show() 
```

##### Using technical analysis library 'ta'

```python
# MACD is a type of momentum indicators which is differnce between two ema 12 and 26
x_train['macd']=ta.trend.macd(x_train['Adj Close'], n_fast=12, n_slow=26, fillna=True)
x_test['macd']=ta.trend.macd(x_test['Adj Close'], n_fast=12, n_slow=26, fillna=True)
```

```python

# fILLING NAN WITH CORRESPONDING CLOSE PRICE 
x_train['MA3']= x_train['MA3'].replace(np.nan, x_train['Adj Close'])
x_test['MA3']= x_test['MA3'].replace(np.nan, x_test['Adj Close'])
x_train['MA2']= x_train['MA2'].replace(np.nan, x_train['Adj Close'])
x_test['MA2']= x_test['MA2'].replace(np.nan, x_test['Adj Close'])

x_train['MAo3']= x_train['MAo3'].replace(np.nan, x_train['Open'])
x_test['MAo3']= x_test['MAo3'].replace(np.nan, x_test['Open'])
x_train['MAo2']= x_train['MAo2'].replace(np.nan, x_train['Open'])
x_test['MAo2']= x_test['MAo2'].replace(np.nan, x_test['Open'])

x_train=x_train.drop(['Adj Close'],axis=1)
x_test=x_test.drop(['Adj Close'],axis=1)
x_train=x_train.drop(['Close'],axis=1)
x_test=x_test.drop(['Close'],axis=1)

x_train.head()
```

```python
# Normalizing data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

# Part 2:Applying time series model (LSTM)

1.In part1 we have just seen normal regression techniques to predict just the next day price but now we will predict coming n days price by looking previous n-1 days.

2.We will apply LSTM because of its property to remember long term dependencies .

3.Its a many to many architecture model.

```python
#https://github.com/alexavierc/LSTM-Stock-Prices/blob/master/LSTM%20Single%20Company.ipynb
```

```python
# preparing data suitable for inputing to lstm
def processData(data,target, look_back, forward_days,jump=1):
    X,Y = [],[]
    for i in range(0,len(data) -look_back-forward_days +1, jump):
        X.append(data[i:(i+look_back)])
        Y.append(target[(i+look_back):(i+look_back+forward_days)])
    return np.array(X),np.array(Y)
```

```python
X_test,y_test = processData(x_test,y_test,20,5) #LOOK_BACK =20 which means we are considering previous 20 days data to predict/ 
y_test = np.array([list(a.ravel()) for a in y_test]) # future (forward_days) 5 days closing price
X_train,y_train = processData(x_train,y_train,20,5)
y_train = np.array([list(a.ravel()) for a in y_train])
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### LSTM model

```python
from keras.layers import Input, Dense
from keras.models import Model
from keras.layers import LSTM,CuDNNLSTM

NUM_NEURONS_FirstLayer = 50
NUM_NEURONS_SecondLayer = 50


#Build the model
model = Sequential()
model.add(CuDNNLSTM(NUM_NEURONS_FirstLayer,input_shape=(20,15), return_sequences=True))
model.add(CuDNNLSTM(NUM_NEURONS_SecondLayer,input_shape=(NUM_NEURONS_FirstLayer,1)))
model.add(Dense(64))
model.add(Dense(5))
model.compile(loss='mean_squared_error', optimizer='adam')

history = model.fit(X_train,y_train,epochs=30,shuffle=True,batch_size=2, verbose=1)
```

```python
Xt = model.predict(X_test)

plt.figure(figsize = (15,10))

for i in range(0,140):
    plt.plot([x + i for x in range(len(Xt[i]))], (Xt[i].reshape(-1,1)), color='r')
    plt.plot([x + i for x in range(len(y_test[i]))], (y_test[i].reshape(-1,1)), color='b')
plt.plot(0, (Xt[i].reshape(-1,1))[0], color='r', label='Prediction') #only to place the label
plt.plot(0, (y_test[i].reshape(-1,1))[0],label='autual',color='b')    
#plt.plot((y_test[:,0]), label='Target')

plt.show()
```

## CONCLUSION

1.Its very hard to predict future prices just based on historical price data because prices are just results of demand and supply between traders and stock prices are dependent on many other factors.

2.Since we are only limited to day prices, if we had hourly or minutes data the prediction may have been some what better.

3.We have seen in ML model that how giving previous day data to predict today price just predicted the same days data .
